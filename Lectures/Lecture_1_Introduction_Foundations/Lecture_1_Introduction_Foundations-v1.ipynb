### Some Simple Examples

#### Find People by Last Name and Birth City

````python
def find_people_by_lastname_and_city(nameLast, birthCityName):
    
    result = []
    
    with open (lahman_dir + "\\People.csv", "r") as people_csv:
        people_dict_rdr = csv.DictReader(people_csv)
        for r in people_dict_rdr:
            if r['nameLast'] == nameLast and r['birthCity'] == birthCityName:
                result.append(r)
            
    return result
````

````python
the_people = find_people_by_lastname_and_city("Williams", "San Diego")
the_people
````

````python
def sql_people_by_name_city(nameLast, birthCityName):
    result = %sql select * from lahmansbaseballdb.people where nameLast = :nameLast and birthCity = :birthCityName
    return result
````

````python
some_people = sql_people_by_name_city('Williams', 'San Diego')
some_people
````

- We can argue about which one is "simpler," but is one of them faster?

````python
import time

start_t = time.time()
for i in range(0,10):
    result = find_people_by_lastname_and_city("Williams", "San Diego")
end_t = time.time()
file_t = end_t - start_t

print("The elapsed time to get the answer with CSV file is:", file_t)
````

````python
start_t = time.time()
for i in range(0,10):
    result = sql_people_by_name_city('Williams', 'San Diego')
end_t = time.time()
sql_t = end_t - start_t

print("The elapsed time to get the answer with a database is:", sql_t)
````

````python
print("\nBy this simple example, SQL is", round(file_t/sql_t, 1), "fast than files.")
````

- In general, databases engines:
    - Implement special libraries and languages that vastly simplify working with complex data.
    - Perform operations __exponentially__ faster than manipulating files, using spreadsheets, ... ...

## Overview

- There is a website and project with some interesting information from the TV Series

````python
import pandas
````

````python
l_data = %sql SELECT location, subLocation, count(*) as count  FROM gotprocessed.scenes \
	group by location, subLocation having count > 10 order by count desc;
````

````python
l_df = l_data.DataFrame()
l_df = l_df[['location', 'subLocation', 'count']]
````

````python
l_df.head(10)
````

````python
import seaborn as sb
import matplotlib.pyplot as plt
````

````python
heatmap1_data = pandas.pivot_table(l_df, values='count', 
                     index=['location'], 
                     columns='subLocation')
````

````python
plt.figure(figsize=(20, 10))
sb.heatmap(heatmap1_data, cmap="YlGnBu")

plt.show()
````
